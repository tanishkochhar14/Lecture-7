```python
# Report_Card_Dictionary

report_card_dictionary = {"student_names": ["Harsha", "Ashwini", "Rohit", "Mahir"],
                          "math_scores": [50,75,65,70],
                          "chemistry_scores": [80,75,90,65],
                          "information_technology_scores": [40,45,50,65],
                          "physics_scores": [80,60,50,45]}
report_card_dictionary
```

```python
# Setup
import pandas as pd
```

```python
# Converting Dictionary to DataFrame
report_card_df = pd.DataFrame(report_card_dictionary)
report_card_df
```

```python
report_card_df.head()
```

```python
# First two rows in the database.
report_card_df.head(2)
```

```python
# Sample of two rows in our data frame 
report_card_df.sample(3)
```

```python
# Create a description of our numeric columns
report_card_df.describe()
```

```python
# Create a column called total score.
report_card_df["total_score"] = report_card_df["math_scores"] + report_card_df["chemistry_scores"] + report_card_df["information_technology_scores"] + report_card_df["physics_scores"]
report_card_df.head(5)
```

#Challenge: Create a column called average score


```python
report_card_df["avg_score"] = report_card_df["total_score"]/4
report_card_df
```
